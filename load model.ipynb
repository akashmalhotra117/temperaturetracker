```python
sc_X = pickle.load(open('sc_X3.pkl','rb'))
sc_Y = pickle.load(open('sc_Y3.pkl','rb'))
model = keras.models.load_model('C://Users/Akash/New folder (4)/Untitled Folder/model3')
model.load_weights('weights_realfeel3.h5')
path = 'C://Users/Akash/New folder (4)/Untitled Folder/north/'
all_filenames = []
cities = ["bangalore", "chennai", "hyderabad", "thiruvananthapuram"]
for city in cities:
#     data_files = 'C://Users/Akash/New folder (4)/Untitled Folder/north/*2021.csv'
# all_filenames = [i for i in glob.glob(f"*{file_extension}")]
    all_filenames.append(path+city+'-2018.csv')
for file in all_filenames:
    df = pd.read_csv(file)
    df['newyear'] = pd.to_datetime('2018/01/01')
    df['date']=pd.to_datetime(df['date'])
    df['days']=df['date'] - df['newyear']
    df['days'] = df['days'].dt.days
    df.rename(columns={'maxwspd':'windspd','maxtemp':'tmax', 'mintemp':'tmin', 'min humidity':'prain'}, inplace= True)
    df = df[df['tmin']> 0]
    X = df[['tmin']]
    
#     X= df.where(['tmin'>0])
    col_names = list(X.columns)
#     sc_X = pickle.load(open('sc_X2.pkl','rb'))
#     sc_Y = pickle.load(open('sc_Y2.pkl','rb'))
    X_df= sc_X.fit_transform(X)
    X_df = pd.DataFrame(X_df, columns=col_names)
    y = model.predict(X_df)
    realfeelmin = sc_Y.inverse_transform(y)
    df['realfeelmin']=realfeelmin + 0.75
    df.to_csv(file)
#     X_df.describe().T
```
