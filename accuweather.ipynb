```python
from bs4 import BeautifulSoup as soup
import requests
import pandas as pd
import time
import random
import re
from datetime import date
today = date.today()
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.164 Safari/537.36'}
tmax = []
realfeelmax = []
realfeel_shade= []
windspd=[]
winddir = []
windgust = []
prain =[]
ppt =[]
tmin =[]
realfeelmin = []
date = []
cities = ['mumbai/204842','delhi/202396','bengaluru/204108','hyderabad/202190','ahmedabad/202438','chennai/206671','kolkata/206690','surat/202441','pune/204848','jaipur/205617','lucknow/206678','kanpur/206679','nagpur/204844','indore/204411','bhopal/204408','visakhapatnam/202192','patna/202349','vadodara/202439','ghaziabad/206683','ludhiana/205592','agra/206686','nashik/189304','ranchi/203541','faridabad/202446','meerut/206682','rajkot/202440','varanasi/206681','srinagar/361707','aurangabad/189320','dhanbad/188659','amritsar/205593','allahabad/3114621','gwalior/204409','jabalpur/204410','coimbatore/206673','jodhpur/205618','madurai/206674','raipur/187458','chandigarh/202350','guwahati/186893','solapur/204849','hubli-dharwad/3352291','mysore/204111','tiruchirappalli/190803','bareilly/206684','aligarh/206685','tiruppur/4-190801_1_al','gurgaon/188408','jalandhar/195139','moradabad/191056','bhubaneswar/189781','salem/190799','warangal/202191','jalgaon/189308','kota/190216','guntur/186787','saharanpur/191026','gorakhpur/191048','bikaner/190207','amravati/189309','noida/3146227','jamshedpur/203542','bhilai-nagar/187461','cuttack/189792','firozabad/191034','kochi/204289','nellore/186822','bhavnagar/188166','dehradun/191339','jammu/361708','thiruvananthapuram/204287','agartala/190964','aizawl/189561','imphal/189551','pondicherry/190056','gandhinagar/188134','shimla/188519','amaravati/2942921','gangtok/190542','']
# cities = ['mumbai/204842','delhi/202396']
for city in cities:
    tmax = []
    realfeelmax = []
    realfeel_shade= []
    windspd=[]
    winddir = []
    windgust = []
    prain =[]
    ppt =[]
    tmin =[]
    realfeelmin = []
    date = []
    for i in range(2, 92):
        main_url = "https://www.accuweather.com/en/in/"+ city +"/daily-weather-forecast/" +city.split('/')[1]+"?day=" + str(i)
        pageopen = requests.get(main_url, headers=headers)
        mainhtml=pageopen.text
        pageopen.close()
        mainparsedpg = soup(mainhtml, 'html.parser')
    #     date =[]
        date.append(mainparsedpg.find("div", {"class": "subnav-pagination"}).text.strip() + ', 2021')
        s = mainparsedpg.find_all("div", {"class":"real-feel"})[0].text
        tempmax = mainparsedpg.find_all("div", {"class":"temperature"})[0].text
        s = re.sub('\s+', '', s)
    #     realfeelmax.append(s[9:11])
    #     realfeel_shade = s[26:28]
        s = re.sub('\s+', '', s)
        realfeelmax.append(s[9:11])
        realfeel_shade.append(s[26:28])
        tempmax = re.sub('\s+', '', tempmax)
        tmax.append(tempmax[:2])
        panelitems = mainparsedpg.find_all("p", {"class":"panel-item"})
        windspeed = panelitems[1].find("span", {"class":"value"}).text
        windspd.append(re.sub("[^0-9]", "", windspeed))
        winddirection = panelitems[1].find("span", {"class":"value"}).text[:3]
        winddir.append(re.sub('[^a-zA-Z]+', '', winddirection))
        windgusty = panelitems[2].find("span", {"class":"value"}).text
        windgust.append(re.sub("[^0-9]", "", windgusty))
        prain.append(panelitems[3].find("span", {"class":"value"}).text)
        pptn = panelitems[5].find("span", {"class":"value"}).text
        ppt.append(re.findall("\d+\.\d+",pptn)[0])

        tempmin = mainparsedpg.find_all("div", {"class":"temperature"})[1].text
        tempmin = (re.sub('\s+', '', tempmin))
        tmin.append(re.sub("[^0-9]", "", tempmin))
        realfeel_min = mainparsedpg.find_all("div", {"class":"real-feel"})[1].text
        realfeelmin.append(re.sub("[^0-9]", "", realfeel_min))
        print(i)
        time.sleep(random.uniform(0.5,2))

    weather = {"date":date, "tmax":tmax, "realfeelmax":realfeelmax, "realfeel_shade":realfeel_shade, "windspd":windspd, "winddir":winddir, "windgust":windgust, "prain":prain, "ppt":ppt, "tmin":tmin, "realfeelmin":realfeelmin}
    df = pd.DataFrame(weather)
    df['date']= pd.to_datetime(df['date'])
    df.to_csv(city.split('/')[0] + ".csv")
    time.sleep(random.uniform(0.5,2.5))
```
